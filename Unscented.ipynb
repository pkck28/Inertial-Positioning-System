```python
from filterpy.kalman import UnscentedKalmanFilter
from filterpy.kalman import MerweScaledSigmaPoints
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/home/pavan/Inertial-Positioning-System/data_1.csv")
zs = np.asanyarray(data.GyroX)
data = data.reset_index()
dt = 0.06
```

```python
def fx(x,dt):
    x = x
    return x

def hx(x):
    x = x
    return x
```

```python
points = MerweScaledSigmaPoints(2, alpha=.1, beta=2., kappa=1)
```

```python
ukf = UnscentedKalmanFilter(dim_x=2, dim_z=1, fx=fx, hx=hx, points=points, dt=dt)
ukf.Q = np.array([[0.1,0],[0,0.1]])
ukf.R = 200
ukf.P = np.eye(2)*500
mu,cov = ukf.batch_filter(zs)
plt.plot(data.index,zs,'r')
plt.plot(data.index,mu,'b')
```
